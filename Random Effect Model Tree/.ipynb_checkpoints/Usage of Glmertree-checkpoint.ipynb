```r
# just use treatment as regressor
# here the regress_variable (treatment) and the split_var (age,duration&anxiety)
# are different. But they can be the same.
lmm_tree <- lmertree(depression ~ treatment | cluster |+    age 
                     + duration + anxiety, data = DepressionDemo)

lmm_tree
```

# Residuals analysis

```r
 plot(factor(DepressionDemo$cluster), residuals(lmm_tree))
```

```r
scatter.smooth(predict(lmm_tree),  residuals(lmm_tree))
```

# Extract splitting features used in the tree

```r
## fit normal linear regression LMM tree for continuous outcome
lt <- lmertree(depression ~ treatment | cluster | age + anxiety + duration,
               data = DepressionDemo)

##
## list.rules() is an internal, unexported function from package pre:
##
## Arguments:
##
## x required tree object.
##
## i optional argument, allows for specifying a specific node,
##      then the path to that node will be returned.
##
## removecomplements optional argument, if TRUE, perfectly
##      negatively correlated rules are omitted.
##
paths <- pre:::list.rules(lt$tree, removecomplements = FALSE)
vnames <- names(DepressionDemo)
vnames[sapply(sapply(vnames, FUN = function(var) grep(paste(paste(var,"<="),"|",paste(var,">"),sep=""), paths)), length) > 0]
```

# To increase or decrease the number of depth of tree

```r
lmm_tree <- lmertree(depression ~ treatment | cluster |+    age 
                     + duration + anxiety, data = DepressionDemo)
```

```r
## Restrict tree size to smaller value:
smallertree <- lmertree(depression ~ treatment | cluster | age + duration +
                          anxiety, data = DepressionDemo, maxdepth = 2L)
plot(smallertree)
```

```r
## to increase tree size, need to increase the significance level alpha,
## which is employed for the parameter stability tests:
largertree <- lmertree(depression ~ treatment | cluster | age + duration + anxiety,
                 data = DepressionDemo, alpha = .99)
plot(largertree)
```

```r
regress_var = "x"
split_var = paste("x", 1:3, sep="")
cluster_var = "patient"
# the formula 
Formula = as.formula(paste("y~",regress_var,"|",cluster_var,"|",
                          paste(split_var,collapse = "+")))
Formula
```

```r
system.time({
    lmm_tree <- lmertree(Formula, data = DepressionDemo)
})
plot(lmm_tree)
```
